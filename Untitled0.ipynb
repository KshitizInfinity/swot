```python
import numpy as np

def conv2d(image, kernel, stride=1, padding=0):

    if padding > 0:
        image = np.pad(image,
                       ((padding, padding), (padding, padding)),
                       mode='constant')

    img_h, img_w = image.shape
    k_h, k_w = kernel.shape
    out_h = (img_h - k_h) // stride + 1
    out_w = (img_w - k_w) // stride + 1
    output = np.zeros((out_h, out_w))

    for i in range(out_h):
        for j in range(out_w):

            patch = image[i*stride:i*stride+k_h,
                          j*stride:j*stride+k_w]

            output[i, j] = np.sum(patch * kernel)

    return output
```

```python
# Given image
image = np.array([
    [3,1,0,2,4],
    [1,5,3,2,1],
    [0,2,6,4,3],
    [2,3,1,5,2],
    [1,0,2,3,4]
])

# Sobel-X kernel
kernel = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]
])

output = conv2d(image, kernel, stride=1, padding=0)

print("Output:\n", output)
print("Output shape:", output.shape)
```

## Problem 4

Input (32×32×3)

[Block 1]
Conv2D(32) → BatchNorm → ReLU → MaxPool

[Block 2]
Conv2D(64) → BatchNorm → ReLU → MaxPool

[Block 3]
Conv2D(128) → BatchNorm → ReLU → MaxPool

GlobalAveragePooling

Dense(128) → ReLU → Dropout(0.5)

Dense(10) → Softmax

The model uses three convolutional blocks to progressively extract features from low-level patterns (edges, textures) to high-level representations. Batch Normalization is used to stabilize training, while ReLU introduces non-linearity. MaxPooling reduces spatial dimensions and computation. GlobalAveragePooling reduces parameters compared to Flatten and prevents overfitting. Dropout is added to improve generalization.

```python
from tensorflow.keras import layers, models

model = models.Sequential([

    # Block 1
    layers.Conv2D(32, (3,3), padding='same', input_shape=(32,32,3)),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.MaxPooling2D((2,2)),

    # Block 2
    layers.Conv2D(64, (3,3), padding='same'),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.MaxPooling2D((2,2)),

    # Block 3
    layers.Conv2D(128, (3,3), padding='same'),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.MaxPooling2D((2,2)),

    # 🔥 Extra Conv Layer (key fix)
    layers.Conv2D(256, (3,3), padding='same'),
    layers.BatchNormalization(),
    layers.ReLU(),

    # Global pooling
    layers.GlobalAveragePooling2D(),

    # Dense head
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.5),

    layers.Dense(10, activation='softmax')
])

model.summary()
```

```python
from tensorflow.keras import layers, models

def create_lenet():
    model = models.Sequential([
        layers.Conv2D(6, (5,5), activation='tanh', input_shape=(28,28,1)),
        layers.AveragePooling2D((2,2)),

        layers.Conv2D(16, (5,5), activation='tanh'),
        layers.AveragePooling2D((2,2)),

        layers.Flatten(),
        layers.Dense(120, activation='tanh'),
        layers.Dense(84, activation='tanh'),
        layers.Dense(10, activation='softmax')
    ])
    return model
```

```python
from tensorflow.keras.optimizers import SGD, Adam

histories = {}

# 1. SGD
model = create_lenet()
model.compile(optimizer=SGD(learning_rate=0.01),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

histories['SGD'] = model.fit(
    x_train_mnist_p, y_train_mnist_p,
    epochs=15, batch_size=64,
    validation_split=0.1,
    verbose=0
)

# 2. SGD + Momentum
model = create_lenet()
model.compile(optimizer=SGD(learning_rate=0.01, momentum=0.9),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

histories['Momentum'] = model.fit(
    x_train_mnist_p, y_train_mnist_p,
    epochs=15, batch_size=64,
    validation_split=0.1,
    verbose=0
)

# 3. Adam
model = create_lenet()
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

histories['Adam'] = model.fit(
    x_train_mnist_p, y_train_mnist_p,
    epochs=15, batch_size=64,
    validation_split=0.1,
    verbose=0
)
```

```python
def create_cifar_model():
    from tensorflow.keras import layers, models

    model = models.Sequential([
        layers.Conv2D(32, (3,3), padding='same', input_shape=(32,32,3)),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.MaxPooling2D((2,2)),

        layers.Conv2D(64, (3,3), padding='same'),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.MaxPooling2D((2,2)),

        layers.Conv2D(128, (3,3), padding='same'),
        layers.BatchNormalization(),
        layers.ReLU(),
        layers.MaxPooling2D((2,2)),

        layers.Conv2D(256, (3,3), padding='same'),
        layers.BatchNormalization(),
        layers.ReLU(),

        layers.GlobalAveragePooling2D(),

        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),

        layers.Dense(10, activation='softmax')
    ])
    return model
```

```python
import numpy as np
from tensorflow.keras.datasets import cifar10

(x_train_cifar, y_train_cifar), (x_test_cifar, y_test_cifar) = cifar10.load_data()

def preprocess_data(images, labels, dataset_name="MNIST"):
    print("\nBefore Processsing")
    print("Shape:", images.shape)
    print("Dtype:", images.dtype)
    print("Min:", images.min(), "Max:", images.max())
    print("Sample labels:", labels[:5])

    # Normalization
    images = images.astype('float32') / 255.0

    print("\n After Normalisation")
    print("Dtype:", images.dtype)
    print("Min:", images.min(), "Max:", images.max())

    # Reshape for MNIST
    if dataset_name == "MNIST":
        images = images.reshape(images.shape[0], 28, 28, 1)
        print("\n After Reshaping")
        print("New shape:", images.shape)

    # One-hot encoding
    one_hot_labels = np.zeros((labels.shape[0], 10))
    for i, label in enumerate(labels):
        one_hot_labels[i, label] = 1

    print("\n After One hot encoding")
    print("Label shape:", one_hot_labels.shape) # Fixed typo here
    print("Sample encoded labels:\n", one_hot_labels[:5])

    return images, one_hot_labels

x_train_cifar_p, y_train_cifar_p = preprocess_data(
    x_train_cifar, y_train_cifar.flatten(), "CIFAR"
)
```

```python
from tensorflow.keras.optimizers import SGD

learning_rates = [0.1, 0.01, 0.001]
batch_sizes = [32, 128]

results = {}

for lr in learning_rates:
    for bs in batch_sizes:
        print(f"\nTraining with LR={lr}, Batch Size={bs}")

        model = create_cifar_model()

        model.compile(
            optimizer=SGD(learning_rate=lr),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        history = model.fit(
            x_train_cifar_p, y_train_cifar_p,
            epochs=10,
            batch_size=bs,
            validation_split=0.1,
            verbose=0
        )

        final_val_acc = history.history['val_accuracy'][-1]

        results[(lr, bs)] = final_val_acc
```

```python
def create_model(reg_type="none"):
    from tensorflow.keras import layers, models

    model = models.Sequential()

    # Block 1
    model.add(layers.Conv2D(32, (3,3), padding='same', input_shape=(32,32,3)))

    if reg_type in ["batchnorm", "both"]:
        model.add(layers.BatchNormalization())

    model.add(layers.ReLU())
    model.add(layers.MaxPooling2D((2,2)))

    if reg_type in ["dropout", "both"]:
        model.add(layers.Dropout(0.3))


    # Block 2
    model.add(layers.Conv2D(64, (3,3), padding='same'))

    if reg_type in ["batchnorm", "both"]:
        model.add(layers.BatchNormalization())

    model.add(layers.ReLU())
    model.add(layers.MaxPooling2D((2,2)))

    if reg_type in ["dropout", "both"]:
        model.add(layers.Dropout(0.3))


    # Head
    model.add(layers.GlobalAveragePooling2D())

    model.add(layers.Dense(128, activation='relu'))

    if reg_type in ["dropout", "both"]:
        model.add(layers.Dropout(0.5))

    model.add(layers.Dense(10, activation='softmax'))

    return model
```

```python
from tensorflow.keras.optimizers import Adam

histories = {}

configs = ["none", "dropout", "batchnorm", "both"]

for cfg in configs:
    print(f"\nTraining: {cfg}")

    model = create_model(cfg)

    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    history = model.fit(
        x_train_cifar_p, y_train_cifar_p,
        epochs=20,
        batch_size=64,
        validation_split=0.1,
        verbose=0
    )

    histories[cfg] = history
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

for name, history in histories.items():
    plt.plot(history.history['val_accuracy'], label=name)

plt.title("Regularisation Comparison")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.legend()

plt.savefig("regularisation_comparison.png")
plt.show()
```

```python
import tensorflow as tf

class LRTracker(tf.keras.callbacks.Callback):
    def __init__(self):
        self.lrs = []

    def on_epoch_end(self, epoch, logs=None):
        lr = self.model.optimizer.learning_rate

        # Handle schedule vs constant LR
        if callable(lr):
            lr = lr(self.model.optimizer.iterations)

        self.lrs.append(float(tf.keras.backend.get_value(lr)))
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau

model_plateau = create_model("both")

model_plateau.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    verbose=0
)

lr_tracker_plateau = LRTracker()

history_plateau = model_plateau.fit(
    x_train_cifar_p, y_train_cifar_p,
    epochs=30,
    batch_size=64,
    validation_split=0.1,
    callbacks=[reduce_lr, lr_tracker_plateau],
    verbose=0
)
```

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# 1. SETUP DEVICE
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 2. DATA PREPARATION (CIFAR-10)
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Load dataset
cifar_data = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
train_loader = DataLoader(cifar_data, batch_size=64, shuffle=True)

# 3. DEFINE MODEL (Custom CIFAR-10 CNN)
class ScaledCIFAR10Net(nn.Module):
    def __init__(self):
        super(ScaledCIFAR10Net, self).__init__()
        self.features = nn.Sequential(
            # Block 1: Input (3, 32, 32) -> Output (64, 16, 16)
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),

            # Block 2
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),

            # Block 3
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2, 2)
        )
        self.head = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Dropout(0.4),
            nn.Linear(256, 10)
        )

    def forward(self, x):
        return self.head(self.features(x))

# 4. INSTANTIATE MODEL
model = ScaledCIFAR10Net().to(device)

# 5. FILTER VISUALIZATION FUNCTION
def visualize_conv1_filters(model, filename="conv1_filters.png"):
    model.eval()
    # Extract weights from the first Conv2d layer
    # model.features[0] is the first Conv2d(3, 64, ...)
    weights = model.features[0].weight.data.cpu().numpy()

    fig, axes = plt.subplots(8, 8, figsize=(12, 12))
    for i in range(64):
        ax = axes[i // 8, i % 8]

        # Get filter i, shape [3, 3, 3]
        f = weights[i]

        # Normalise to [0, 1]
        f_min, f_max = f.min(), f.max()
        f = (f - f_min) / (f_max - f_min + 1e-5)

        # Reorder dimensions for plotting (C, H, W) -> (H, W, C)
        f = np.transpose(f, (1, 2, 0))

        ax.imshow(f)
        ax.set_title(f"Idx: {i}", fontsize=8)
        ax.axis('off')

    plt.tight_layout()
    plt.savefig(filename)
    plt.show()
    print(f"Filters saved as {filename}")

# Run the visualization
visualize_conv1_filters(model)
```

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.gradients = None
        self.activations = None

        self.target_layer.register_forward_hook(self.save_activation)
        self.target_layer.register_full_backward_hook(self.save_gradient)

    def save_activation(self, module, input, output):
        self.activations = output

    def save_gradient(self, module, grad_input, grad_output):
        self.gradients = grad_output[0]

    def generate(self, x, class_idx=None):
        preds = self.model(x)
        if class_idx is None:
            class_idx = preds.argmax(dim=1).item()

        self.model.zero_grad()
        score = preds[0, class_idx]
        score.backward(retain_graph=True)

        pooled_grads = torch.mean(self.gradients, dim=[0, 2, 3])

        acts = self.activations[0].detach()
        for i in range(acts.size(0)):
            acts[i, :, :] *= pooled_grads[i]

        heatmap = torch.sum(acts, dim=0).cpu().numpy()
        heatmap = np.maximum(heatmap, 0)

        max_val = np.max(heatmap)
        if max_val > 0:
            heatmap /= max_val

        return heatmap, preds.argmax(dim=1).item()

def overlay_heatmap(img_tensor, heatmap):
    img = img_tensor.detach().cpu().numpy().transpose(1, 2, 0)
    img = img * 0.5 + 0.5
    img = np.clip(img, 0, 1)

    heatmap_img = Image.fromarray(np.uint8(255 * heatmap))
    heatmap_resized = heatmap_img.resize((32, 32), Image.Resampling.BILINEAR)
    heatmap_resized = np.array(heatmap_resized) / 255.0

    cmap = plt.get_cmap('jet')
    heatmap_colored = cmap(heatmap_resized)[:, :, :3]

    superimposed = 0.4 * heatmap_colored + 0.6 * img
    return np.clip(superimposed, 0, 1)

model.eval()
for param in model.parameters():
    param.requires_grad = True

target_layer = model.features[8]
cam = GradCAM(model, target_layer)

correct_images = []
misclassified_image = None

for img, label in test_loader:
    img, label = img.to(device), label.to(device)
    img.requires_grad = True

    out = model(img)
    pred = out.argmax(dim=1).item()
    true = label.item()

    if pred == true and len(correct_images) < 3:
        correct_images.append((img, true, pred))
    elif pred != true and misclassified_image is None:
        misclassified_image = (img, true, pred)

    if len(correct_images) == 3 and misclassified_image is not None:
        break

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, (img, true_label, pred_label) in enumerate(correct_images):
    heatmap, _ = cam.generate(img, class_idx=true_label)
    overlay = overlay_heatmap(img.squeeze(), heatmap)

    axes[i].imshow(overlay)
    axes[i].set_title(f"Correct: {classes[true_label]}")
    axes[i].axis('off')

img, true_label, pred_label = misclassified_image

heatmap_true, _ = cam.generate(img, class_idx=true_label)
overlay_true = overlay_heatmap(img.squeeze(), heatmap_true)
axes[3].imshow(overlay_true)
axes[3].set_title(f"Misclassified\nTrue Class: {classes[true_label]} (Heatmap)")
axes[3].axis('off')

heatmap_pred, _ = cam.generate(img, class_idx=pred_label)
overlay_pred = overlay_heatmap(img.squeeze(), heatmap_pred)
axes[4].imshow(overlay_pred)
axes[4].set_title(f"Misclassified\nPred Class: {classes[pred_label]} (Heatmap)")
axes[4].axis('off')

axes[5].axis('off')

plt.tight_layout()
plt.savefig("gradcam_results.png", dpi=300)
plt.show()
print("Saved as gradcam_results.png")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader

# 1. Setup Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# 2. Data Preparation (ImageNet Preprocessing)
# CIFAR-10 resized to 96x96 and normalized using ImageNet mean/std
transform = transforms.Compose([
    transforms.Resize((96, 96)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_data = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
val_data = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_data, batch_size=128, shuffle=True)
val_loader = DataLoader(val_data, batch_size=128, shuffle=False)

# 3. Load Pre-trained ResNet50
# Using the recommended standard weights
model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)

# (a) Freeze all convolutional weights
for param in model.parameters():
    param.requires_grad = False

# Add custom classification head
# ResNet50 already has a Global Average Pooling layer (model.avgpool).
# We just need to replace the final Fully Connected (fc) layer.
in_features = model.fc.in_features # For ResNet50, this is 2048

model.fc = nn.Sequential(
    nn.Linear(in_features, 256),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(256, 10)
    # Softmax is omitted here because CrossEntropyLoss handles it natively
)

model = model.to(device)

# (c) Calculate Trainable vs Frozen Parameters
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
frozen_params = sum(p.numel() for p in model.parameters() if not p.requires_grad)

print("-" * 40)
print(f"Frozen parameters (Conv base): {frozen_params:,}")
print(f"Trainable parameters (New Head): {trainable_params:,}")
print("-" * 40)

# 4. Training Setup
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

history = {'train_acc': [], 'val_acc': []}
epochs = 10

# 5. Training Loop
for epoch in range(epochs):
    # Train
    model.train()
    correct_train, total_train = 0, 0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        correct_train += (outputs.argmax(dim=1) == labels).sum().item()
        total_train += labels.size(0)

    train_acc = correct_train / total_train
    history['train_acc'].append(train_acc)

    # Validation
    model.eval()
    correct_val, total_val = 0, 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            correct_val += (outputs.argmax(dim=1) == labels).sum().item()
            total_val += labels.size(0)

    val_acc = correct_val / total_val
    history['val_acc'].append(val_acc)

    print(f"Epoch [{epoch+1}/{epochs}] | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")

# 6. Plotting
plt.figure(figsize=(8, 6))
plt.plot(range(1, epochs+1), history['train_acc'], label='Train Accuracy', marker='o')
plt.plot(range(1, epochs+1), history['val_acc'], label='Validation Accuracy', marker='o')
plt.title('Transfer Learning: ResNet50 Frozen Base')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('tl_frozen.png', dpi=300)
plt.show()
print("Saved plot to tl_frozen.png")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import copy

def count_trainable_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

def setup_model(unfreeze_mode):
    # 1. Load fresh pre-trained model and freeze everything
    model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
    for param in model.parameters():
        param.requires_grad = False

    # 2. Add custom head (always unfreezing the head)
    model.fc = nn.Sequential(
        nn.Linear(2048, 256),
        nn.ReLU(),
        nn.Dropout(0.5),
        nn.Linear(256, 10)
    )

    # 3. Apply Unfreezing Strategy
    if unfreeze_mode == 'ALL':
        for param in model.parameters():
            param.requires_grad = True
    elif unfreeze_mode > 0:
        # Extract all Conv2d layers
        conv_layers = [m for m in model.modules() if isinstance(m, nn.Conv2d)]
        target_convs = conv_layers[-unfreeze_mode:]

        # Unfreeze target Convs
        for conv in target_convs:
            for param in conv.parameters():
                param.requires_grad = True

        # Best Practice: Also unfreeze the Batch Norm layers in the final blocks
        # For simplicity in this loop, we unfreeze all BNs in layer4 if unfreezing > 0
        for param in model.layer4.parameters():
            # If a parameter requires grad, keep it true. Otherwise, check if it's BN.
            pass # (In a stricter implementation, we explicitly target matching BNs)

        # A safer, block-based unfreezing for ResNet50 top layers:
        if unfreeze_mode == 2:
            for param in model.layer4[2].parameters(): param.requires_grad = True
        elif unfreeze_mode == 8:
            for param in model.layer4.parameters(): param.requires_grad = True

    return model.to(device)

def run_experiment(unfreeze_mode):
    print(f"\n{'='*40}\nRunning Experiment: Unfreeze {unfreeze_mode}\n{'='*40}")
    model = setup_model(unfreeze_mode)

    # Using 1e-5 as requested for fine-tuning
    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-5)
    criterion = nn.CrossEntropyLoss()

    trainable_params = count_trainable_params(model)
    print(f"Trainable Parameters: {trainable_params:,}")

    best_val_acc = 0.0
    final_train_acc = 0.0

    for epoch in range(10): # 10 Fine-tuning epochs
        # --- Training ---
        model.train()
        correct_t, total_t = 0, 0
        for imgs, lbls in train_loader:
            imgs, lbls = imgs.to(device), lbls.to(device)
            optimizer.zero_grad()
            out = model(imgs)
            loss = criterion(out, lbls)
            loss.backward()
            optimizer.step()
            correct_t += (out.argmax(1) == lbls).sum().item()
            total_t += lbls.size(0)
        train_acc = correct_t / total_t

        # --- Validation ---
        model.eval()
        correct_v, total_v = 0, 0
        with torch.no_grad():
            for imgs, lbls in val_loader:
                imgs, lbls = imgs.to(device), lbls.to(device)
                out = model(imgs)
                correct_v += (out.argmax(1) == lbls).sum().item()
                total_v += lbls.size(0)
        val_acc = correct_v / total_v

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            final_train_acc = train_acc # Record train acc at the best val epoch

        print(f"Epoch {epoch+1} | Train: {train_acc:.4f} | Val: {val_acc:.4f}")

    gap = (final_train_acc - best_val_acc) * 100
    overfit = "Yes" if gap > 5.0 else "No"

    return trainable_params, best_val_acc, gap, overfit

# Execute the 3 configurations
results = {}
experiments = [('Top 2', 2), ('Top 8', 8), ('ALL', 'ALL')]

for name, mode in experiments:
    results[name] = run_experiment(mode)
```
